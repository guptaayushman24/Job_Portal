```python
from sentence_transformers import SentenceTransformer

# Load the pre-trained model
model = SentenceTransformer('paraphrase-mpnet-base-v2')
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Define two texts
text1 = "React.js, CSS, HTML"
text2 = "Frontend development with React.js, CSS, and HTML"

# Encode the texts
embedding1 = model.encode(text1)
embedding2 = model.encode(text2)

# Compute cosine similarity
similarity = cosine_similarity([embedding1], [embedding2])
print("Similarity Score:", similarity[0][0])
```

```python
user_skills = "React.js, CSS, HTML"
job_descriptions = [
    "Frontend development with React.js, CSS, and HTML",
    "Backend development using Python and Django",
    "Data science and machine learning using Python"
]

# Encode user skills once
user_embedding = model.encode(user_skills)

# Compute similarity with each job description
for job in job_descriptions:
    job_embedding = model.encode(job)
    similarity = cosine_similarity([user_embedding], [job_embedding])
    print(f"Job: {job}\nSimilarity Score: {similarity[0][0]:.4f}\n")
```

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
nlp = spacy.load('en_core_web_sm')
def preprocess(text) :
  doc = nlp(text)
  no_stop_words = [token.text for token in doc if not token.is_stop and  not token.is_punct]
  return no_stop_words
```

```python
list = preprocess("We are seeking a Machine Learning Engineer to develop and implement machine learning models to solve real-world problems. The role involves creating scalable solutions for data analysis, predictive modeling, and algorithm optimization. Proficiency in Python, Machine Learning frameworks, and Data Science techniques is essential")
print(' '.join(list))
```
